# 阶跃函数

```python
import numpy as np
import matplotlib.pylab as plt
def step_function(x):
    return np.array(x > 0, dtype=int)
x = np.arange(-5.0, 5.0, 0.1)
y = step_function(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # 指定y轴的范围
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Sigmoid

```python
import numpy as np
import matplotlib.pylab as plt
def sigmoid(x):
    return 1 / (1 + np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1) # 指定y轴的范围
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# ReLU

```python
import numpy as np
import matplotlib.pylab as plt
def relu(x):
    return np.maximum(0, x) # maximum函数会从输入的数值中选择较大的那个值进行输出
x = np.arange(-5.0, 5.0, 0.1)
y = relu(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Tahn

```python
import numpy as np
import matplotlib.pylab as plt
def tahn(x):
    return (np.exp(x)-np.exp(-x)) / (np.exp(x) + np.exp(-x))
x = np.arange(-5.0, 5.0, 0.1)
y = tahn(x)
plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

# Leaky ReLU

```python
import numpy as np
import matplotlib.pylab as plt

def leaky_relu(x, alpha=0.1):
    return np.where(x >= 0, x, alpha*x)

x = np.arange(-5.0, 5.0, 0.1)
y = leaky_relu(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
