```python
import torch
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
lr=0.1
X = torch.unsqueeze(torch.linspace(-1, 1,100), dim=1)
Y = 0.5 * X + 0.3*torch.normal(mean=0.0, std=0.1, size=X.size())
print(X.shape,Y.shape)
plt.scatter(X,Y)
plt.show()

# X = np.linspace(-1, 1, 100)
# Y = 0.5 * X + 0.2* np.random.normal(0, 0.05, (100, ))

# #将X，Y转成200 batch大小，1维度的数据
# X=torch.Tensor(X.reshape(100,1))
# Y=torch.Tensor(Y.reshape(100,1))
# plt.scatter(X,Y)
# plt.show()
```

```python
# 神经网络结构

class Net(torch.nn.Module):
    def __init__(self):
        super(Net, self).__init__()
        self.fc1 = torch.nn.Linear(1, 1)

    def forward(self, x):
        x = self.fc1(x)
        return x

net = Net()
optimizer = torch.optim.SGD(net.parameters(), lr)
loss_function = torch.nn.MSELoss()
```

```python
for i in range(100):
    prediction = net(X)
    loss = loss_function(prediction, Y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    if i%10==0:
        print(loss)
```
